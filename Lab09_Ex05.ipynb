## คำถาม

1. Static Method คืออะไร


ลักษณะของ Static Method
ไม่ต้องสร้างอ็อบเจ็กต์: Static method สามารถเรียกใช้งานได้โดยตรงผ่านชื่อคลาส โดยไม่จำเป็นต้องสร้าง instance ของคลาสนั้น

เข้าถึง Static Members เท่านั้น: Static method สามารถเข้าถึง static members (ทั้ง fields และ methods) ของคลาสได้ แต่ไม่สามารถเข้าถึง instance members ได้โดยตรง

ใช้สำหรับฟังก์ชันทั่วไป: มักใช้สำหรับฟังก์ชันที่ไม่จำเป็นต้องอิงกับข้อมูลของ instance เช่น ฟังก์ชันช่วยคำนวณหรือฟังก์ชันที่จัดการข้อมูลทั่วไป

2. ประกาศ static method ในภาษา C# อย่างไร


ประกาศ static method โดยใช้คำว่า static ตามด้วยประเภทข้อมูลของค่าที่คืน (return type) ชื่อเมธอด และพารามิเตอร์ (ถ้ามี)

เรียกใช้ static method โดยใช้ชื่อคลาสตามด้วยชื่อเมธอดโดยไม่ต้องสร้าง instance ของคลาสนั้น










3. เรียกใช้ static method อย่างไร


เพื่อเรียกใช้ static method ใน C# คุณต้องใช้ชื่อคลาสที่มีการประกาศ static method นั้น ตามด้วยเครื่องหมายจุด

4. static method สามารถเข้าถึง instance members (fields, methods) ได้หรือไม่


Static method ไม่สามารถเข้าถึง instance members (fields และ methods) ได้โดยตรง เนื่องจาก static members จะต้องใช้ชื่อคลาสในการเข้าถึง และไม่ได้อิงกับ instance ของคลาสนั้น ๆ หากต้องการเข้าถึง instance members จะต้องมีการสร้าง instance ของคลาสก่อน

5. static method เหมาะกับการใช้งานแบบไหน


เมื่อฟังก์ชันไม่ต้องการข้อมูลเฉพาะจาก instance ของคลาส

เมื่อต้องการให้ฟังก์ชันสามารถเรียกใช้ได้จากทุกที่โดยไม่ต้องสร้าง instance

เมื่อต้องการสร้างฟังก์ชันช่วย (utility functions) ที่ทำงานเกี่ยวกับข้อมูลทั่วไป

6. ข้อดีของการใช้ static method คืออะไรบ้าง


การเข้าถึงง่าย: สามารถเรียกใช้ได้โดยไม่ต้องสร้าง instance ของคลาส

ประหยัดหน่วยความจำ: ไม่ต้องสร้างอ็อบเจ็กต์ใหม่ ช่วยลดการใช้หน่วยความจำ

ความชัดเจน: สามารถระบุว่าเมธอดไม่เปลี่ยนแปลงสถานะของ instance ได้ชัดเจน

ช่วยในการจัดกลุ่มฟังก์ชัน: สามารถจัดกลุ่มฟังก์ชันที่เกี่ยวข้องในคลาสเดียวกัน

7. ข้อเสียของการใช้ static method คืออะไรบ้าง


ไม่สามารถเข้าถึง instance members ได้: ทำให้ไม่สามารถใช้งานข้อมูลเฉพาะของ instance ได้

ยากต่อการทดสอบ: Static methods อาจทำให้การเขียน unit tests เป็นเรื่องยาก เนื่องจากไม่สามารถ mock หรือ stub ได้ง่าย

ขาดความยืดหยุ่น: ไม่สามารถ override หรือ polymorphism เนื่องจากไม่สามารถใช้กับ instance ของคลาส
